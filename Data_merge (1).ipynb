```python
import zipfile
import os

zip_path = "UAVSwarm-dataset-master.zip"
extract_path = "/content/UAVSwarm"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Done extracting.")
```

```python
import pandas as pd
from glob import glob

base_path = "/content/UAVSwarm/UAVSwarm-dataset-master/train"

det_files = glob(os.path.join(base_path, "UAVSwarm-*", "det", "det.txt"))

print(f"Found {len(det_files)} det files")

det_dfs = []
for f in det_files:
    df = pd.read_csv(f, header=None)
    df["sequence"] = os.path.basename(os.path.dirname(os.path.dirname(f)))  # UAVswarm-XX
    det_dfs.append(df)

det_all = pd.concat(det_dfs, ignore_index=True)

det_all.to_csv("/content/combined_det.csv", index=False, header=False)
print("Saved combined_det.csv")
```

```python
gt_files = glob(os.path.join(base_path, "UAVSwarm-*", "gt", "gt.txt"))

print(f"Found {len(gt_files)} gt files")

gt_dfs = []
for f in gt_files:
    df = pd.read_csv(f, header=None)
    df["sequence"] = os.path.basename(os.path.dirname(os.path.dirname(f)))
    gt_dfs.append(df)

gt_all = pd.concat(gt_dfs, ignore_index=True)

gt_all.to_csv("/content/combined_gt.csv", index=False, header=False)
print("Saved combined_gt.csv")
```

```python
import pandas as pd

det = pd.read_csv("combined_det.csv", header=None)
gt  = pd.read_csv("combined_gt.csv", header=None)

# Drop last column from det
det = det.iloc[:, :-1]

# Merge side-by-side
combined = pd.concat([det, gt], axis=1)

# Save without headers
combined.to_csv("combined_det_gt.csv", index=False, header=False)

combined.head()


```
