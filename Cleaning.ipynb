```python
import pandas as pd
import numpy as np

# Load detection file
# Adjust path as needed
df = pd.read_csv(
    "combined_det.csv",
    sep=r"\s+|,",
    header=None,
    engine="python"
)

# Assign column names based on dataset spec
df.columns = [
    "frame",
    "object_id",
    "x",
    "y",
    "width",
    "height",
    "confidence",
    "ignore_1",
    "ignore_2",
    "ignore_3",
    "Swarm"
]

df.head()
```

```python
# Drop rows with missing required fields
required_cols = ["frame", "x", "y", "width", "height", "confidence"]
df = df.dropna(subset=required_cols)
```

```python
df = df[
    (df["width"] > 0) &
    (df["height"] > 0)
]
```

```python
df = df[
    (df["confidence"] >= 0.0) &
    (df["confidence"] <= 1.0)
]
```

```python
df = df[df["frame"] >= 0]
df["frame"] = df["frame"].astype(int)
```

```python
df = df.drop(columns=["object_id", "ignore_1", "ignore_2", "ignore_3","Swarm"])
```

```python
print("Cleaned dataset shape:", df.shape)
df.describe()
```

```python
df.to_csv("detections_cleaned.csv", index=False)
```
